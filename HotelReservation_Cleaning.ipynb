```python
df = pd.read_csv("drive/MyDrive/Datasets/Hotel Reservations.csv", index_col = "Booking_ID")
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df["booking_status"] = df["booking_status"].map({"Canceled" : 1, "Not_Canceled" : 0})
```

```python
ohe = OneHotEncoder()
cat_columns = df.select_dtypes(include = "object").columns
```

```python
encoded_cols = ohe.fit_transform(df[cat_columns])
encoded_cols
```

```python
encoded_labels = []
for col in np.array(ohe.categories_, dtype = np.ndarray).ravel():
  for cat in col:
    encoded_labels.append(str(cat))
encoded_labels
```

```python
encoded_df = pd.DataFrame(encoded_cols.toarray(), columns = encoded_labels)
df_drop = df.drop(cat_columns, axis = 1)
encoded_df = pd.concat([df_drop.reset_index(drop = True), encoded_df.reset_index(drop = True)], axis = 1)
```

```python
encoded_df
```

```python
features = encoded_df.drop("booking_status", axis = 1)
labels = encoded_df["booking_status"]
```

```python
labels
```

```python
scaler = MinMaxScaler()
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = .2)
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
model = SVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
